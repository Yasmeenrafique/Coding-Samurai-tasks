```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
# Count attrition
total_employees = df.shape[0]
left_employees = df[df['Attrition'] == 'Yes'].shape[0]

attrition_rate = (left_employees / total_employees) * 100
print(f"Attrition Rate: {attrition_rate:.2f}%")
```

```python
# Group attrition by department
dept_attrition = df[df['Attrition'] == 'Yes']['Department'].value_counts()

# Plot
dept_attrition.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Attrition by Department")
plt.xlabel("Department")
plt.ylabel("Number of Employees Left")
plt.show()
```

```python
avg_age_left = df[df['Attrition'] == 'Yes']['Age'].mean()
print(f"Average Age of Employees who Left: {avg_age_left:.1f} years")
```

```python
# Group attrition by department
dept_attrition = df[df['Attrition'] == 'Yes']['Gender'].value_counts()
dept_attrition.plot(kind='barh',color='green',edgecolor='black')
plt.title('Gender wise attrition')
plt.xlabel("Number of employees left")
plt.ylabel("Gender")
plt.show()
```
